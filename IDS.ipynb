```python
import pandas as pd
import numpy as np
```

```python
df= pd.read_excel(r'sqli.xlsx', engine='openpyxl')
df
```

```python
 from sklearn.feature_extraction.text import CountVectorizer
df_MD = [row['Query'] for index,row in df.iterrows() if row['Label'] == 'MD']
df_MD
```

```python
vec = CountVectorizer()
```

```python
l=vec.fit_transform(df_MD)
print(l.toarray())
```

```python
 print(vec.get_feature_names())
```

```python
# print(l.toarray())
table_MD=pd.DataFrame(l.toarray(), columns=vec.get_feature_names())
table_MD
```
